```python
import pandas as pd
```

```python
arastirma_komisyonlari = pd.read_csv("arastirma_komisyonlari/arastirma_komisyonlari.csv")
arastirma_komisyonlari_uyeler = pd.read_csv("arastirma_komisyonlari/arastirma_komisyonlari_uyeler.csv")
arastirma_komisyonlari_df = pd.merge(left=arastirma_komisyonlari, right=arastirma_komisyonlari_uyeler, left_on="id", right_on="arastirma_komisyon_id", how="outer").dropna().drop(["komisyon", "url", "arastirma_komisyon_id"], axis=1).reset_index(drop=True)[["unvan", "adi_ve_soyadi", "secim_bolgesi", "donem", "id"]]
arastirma_komisyonlari_final = pd.DataFrame(arastirma_komisyonlari_df.groupby(["adi_ve_soyadi", "secim_bolgesi", "donem"])["id"].count()).reset_index().rename({"id": "komisyon_uyeligi_count"}, axis=1)
arastirma_komisyonlari_final.columns = ['adi_ve_soyadi', 'secim_cevresi', 'donem', 'komisyon_uyeligi_count']
```

```python
ihtisas_komisyonlari = pd.read_csv("ihtisas_komisyonlari/ihtisas_komisyonlari_uyeler.csv")
ihtisas_komisyonlari_final = pd.DataFrame(ihtisas_komisyonlari.groupby(["adi_ve_soyadi", "partisi", "secim_cevresi"])["komisyon_id"].count()).reset_index().rename({"komisyon_id":"komisyon_uyeligi_count"}, axis=1)
```

```python
genel_gorusme = pd.read_csv("genel_gorusme/genel_gorusme_imza_sahipleri.csv")
genel_gorusme["donem"] = genel_gorusme["genel_gorusme_id"].apply(lambda x: str(x)[:2])
genel_gorusme_imza_count = pd.DataFrame(genel_gorusme.groupby(["adi_ve_soyadi", "partisi", "secim_cevresi", "donem"])["genel_gorusme_id"].count()).reset_index().rename({"genel_gorusme_id":"imza_count"}, axis=1)
genel_gorusme_ilk_imza_count = pd.DataFrame(genel_gorusme.groupby(["adi_ve_soyadi", "partisi", "secim_cevresi", "donem"])["ilk_imza"].sum()).reset_index().rename({"ilk_imza":"ilk_imza_count"}, axis=1)
genel_gorusme_final = pd.merge(left=genel_gorusme_imza_count, right=genel_gorusme_ilk_imza_count, how="outer", on=["adi_ve_soyadi", "partisi", "secim_cevresi", "donem"])
```

```python
gensoru = pd.read_csv("gensoru/gensoru_imza_sahipleri.csv")
gensoru["donem"] = gensoru["gensoru_id"].apply(lambda x: str(x)[:2])
gensoru_imza_count = pd.DataFrame(gensoru.groupby(["adi_ve_soyadi", "partisi", "secim_cevresi", "donem"])["gensoru_id"].count()).reset_index().rename({"gensoru_id":"imza_count"}, axis=1)
gensoru_ilk_imza_count = pd.DataFrame(gensoru.groupby(["adi_ve_soyadi", "partisi", "secim_cevresi", "donem"])["ilk_imza"].sum()).reset_index().rename({"ilk_imza":"ilk_imza_count"}, axis=1)
gensoru_final = pd.merge(left=gensoru_imza_count, right=gensoru_ilk_imza_count, how="outer", on=["adi_ve_soyadi", "partisi", "secim_cevresi", "donem"])
```

```python
kanun_teklifleri = pd.read_csv("kanun_teklifleri/kanun_teklifi_imza_sahipleri.csv")
kanun_teklifleri["donem"] = kanun_teklifleri["kanun_teklifi_id"].apply(lambda x: str(x)[:2])
kanun_teklifleri_imza_count = pd.DataFrame(kanun_teklifleri.groupby(["adi_ve_soyadi", "partisi", "secim_cevresi", "donem"])["kanun_teklifi_id"].count()).reset_index().rename({"kanun_teklifi_id":"imza_count"}, axis=1)
kanun_teklifleri_ilk_imza_count = pd.DataFrame(kanun_teklifleri.groupby(["adi_ve_soyadi", "partisi", "secim_cevresi", "donem"])["ilk_imza"].sum()).reset_index().rename({"ilk_imza":"ilk_imza_count"}, axis=1)
kanun_teklifleri_final = pd.merge(left=kanun_teklifleri_imza_count, right=kanun_teklifleri_ilk_imza_count, how="outer", on=["adi_ve_soyadi", "partisi", "secim_cevresi", "donem"])
```

```python
meclis_arastirma = pd.read_csv("meclis_arastirma/meclis_arastirma_imza_sahipleri.csv", dtype="object")
meclis_arastirma["ilk_imza"] = meclis_arastirma["ilk_imza"].astype("bool")
meclis_arastirma["donem"] = meclis_arastirma["meclis_arastirma_id"].apply(lambda x: str(x)[:2])
meclis_arastirma_imza_count = pd.DataFrame(meclis_arastirma.groupby(["adi_ve_soyadi", "partisi", "secim_cevresi", "donem"])["meclis_arastirma_id"].count()).reset_index().rename({"meclis_arastirma_id":"imza_count"}, axis=1)
meclis_arastirma_ilk_imza_count = pd.DataFrame(meclis_arastirma.groupby(["adi_ve_soyadi", "partisi", "secim_cevresi", "donem"])["ilk_imza"].sum()).reset_index().rename({"ilk_imza":"ilk_imza_count"}, axis=1)
meclis_arastirma_final = pd.merge(left=meclis_arastirma_imza_count, right=meclis_arastirma_ilk_imza_count, how="outer", on=["adi_ve_soyadi", "partisi", "secim_cevresi", "donem"])
```

```python
meclis_sorusturma = pd.read_csv("meclis_sorusturma/meclis_sorusturma_imza_sahipleri.csv")
meclis_sorusturma["donem"] = meclis_sorusturma["meclis_sorusturma_id"].apply(lambda x: str(x)[:2])
meclis_sorusturma_imza_count = pd.DataFrame(meclis_sorusturma.groupby(["adi_ve_soyadi", "partisi", "secim_cevresi", "donem"])["meclis_sorusturma_id"].count()).reset_index().rename({"meclis_sorusturma_id":"imza_count"}, axis=1)
meclis_sorusturma_ilk_imza_count = pd.DataFrame(meclis_sorusturma.groupby(["adi_ve_soyadi", "partisi", "secim_cevresi", "donem"])["ilk_imza"].sum()).reset_index().rename({"ilk_imza":"ilk_imza_count"}, axis=1)
meclis_sorusturma_final = pd.merge(left=meclis_sorusturma_imza_count, right=meclis_sorusturma_ilk_imza_count, how="outer", on=["adi_ve_soyadi", "partisi", "secim_cevresi", "donem"])
```
